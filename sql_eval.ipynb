```python
import csv

def extract_column_to_text(input_csv, output_text, column_index):
    with open(input_csv, 'r') as csv_file, open(output_text, 'w') as text_file:
        # Create a CSV reader
        csv_reader = csv.reader(csv_file)

        # Extract the specified column data and write to the text file
        for row in csv_reader:
            if len(row) > column_index:
                text_file.write(row[column_index] + '\n')

# Example usage:
input_csv_file = 'GPT4_few_shot.csv'
output_text_file = 'pred.txt'
column_index_to_extract = 1  # Replace with the index of the desired column (0-based)

extract_column_to_text(input_csv_file, output_text_file, column_index_to_extract)
```

```python
import json

def read_text_file(file_path):
    with open(file_path, "r") as file:
        lines = file.readlines()
    return [line.strip() for line in lines]

def create_json(db_ids, questions, gold_sqls):
    json_data = []
    for db_id, question, gold_sql in zip(db_ids, questions, gold_sqls):
        json_data.append({
            "db_id": db_id,
            "question": question,
            "gold_sql": gold_sql
        })
    return json_data

# Paths to text files
db_id_file = "pred.txt"
question_file = "ques.txt"
gold_sql_file = "gold.txt"

# Read contents of text files
db_ids = read_text_file(db_id_file)
questions = read_text_file(question_file)
gold_sqls = read_text_file(gold_sql_file)

# Create JSON data
json_data = create_json(db_ids, questions, gold_sqls)

# Write data to JSON file
with open("data.json", "w") as json_file:
    json.dump(json_data, json_file, indent=4)

print("JSON file created successfully.")
```
